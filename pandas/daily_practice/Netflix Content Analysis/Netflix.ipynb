```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from pathlib import Path
```

```python
df = pd.read_csv("NetflixData.csv")
```

```python
num_col = df.select_dtypes(include=np.number).columns
print("There is", len(num_col),"number of numerical columns in the dataset. and numerical columns are :", list(num_col))
```

```python
cat_col = df.select_dtypes(include=("object","str")).columns
print("There are", len(cat_col),"number of categorical columns in the dataset. and categorical columns are :", list(cat_col))
```

```python
df[num_col].hist(figsize=(5,4))
plt.tight_layout()
plt.xlabel("Year")
plt.ylabel("Number of Movies/TV shows")
plt.show()
```

```python
for col in cat_col:
    plt.figure()
    df[col].value_counts().plot(kind='pie',autopct='%1.1f%%',figsize=(10,12))
    plt.title(f"pie chart of {col}")
    plt.ylabel("")
    plt.show()
```

```python
# sql_sales_analysis.py

import sqlite3
import pandas as pd

# Create database
conn = sqlite3.connect("sales.db")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE IF NOT EXISTS sales (
    id INTEGER PRIMARY KEY,
    product TEXT,
    price REAL,
    quantity INTEGER
)
""")

# Insert sample data
data = [
    ("Laptop", 800, 5),
    ("Phone", 500, 10),
    ("Tablet", 300, 7),
    ("Laptop", 900, 3),
    ("Phone", 550, 6)
]

cursor.executemany("INSERT INTO sales (product, price, quantity) VALUES (?, ?, ?)", data)
conn.commit()

# SQL Query
query = """
SELECT product,
       SUM(quantity) as total_quantity,
       SUM(price * quantity) as revenue
FROM sales
GROUP BY product
"""

df = pd.read_sql_query(query, conn)

print(df)

conn.close()
```
